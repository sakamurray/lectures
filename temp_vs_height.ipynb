## U.S. Stadard atmosphere 1976


```python
import matplotlib.pyplot as plt
%matplotlib inline

Hb = [0, 11, 20, 32, 47, 51, 71, 84.8520] # km'
T = []
T.append(288.15) # K
T.append(T[0]-6.5*(Hb[1]-Hb[0]))
T.append(T[1])
T.append(T[2]+1.0*(Hb[3]-Hb[2]))
T.append(T[3]+2.8*(Hb[4]-Hb[3]))
T.append(T[4])
T.append(T[5]-2.8*(Hb[6]-Hb[5]))
T.append(T[6]-2.0*(Hb[7]-Hb[6]))

print(T)
plt.grid(b=True)
plt.plot(T, Hb)
plt.xlim(150, 350)
plt.xlabel('Temperature [K]')
plt.ylim(0, 90)
plt.ylabel("Geopotential altitude [km']")
plt.savefig('standardAtmosphere.pdf')
plt.show()

```
